```r
# Sort least to greatest
high.order <- order(education$high)
education.high <- education[high.order,]
```

```r
# Sort greatest to least
high.order <- order(education$high, decreasing=TRUE)
education.high <- education[high.order,]
```

```r
# Plot the High Column (ordered)
plot(education.high$high)
```

```r
# What if you pass three columns?
# Passing multiple columns
# You get a scatter plot matrix
plot(education[,2:4])

# You get a scatter plot matrix.
# ObserveThe bachelor degree and advanced degree rates are strongly correlated.
```

# Passing Arguments in chart

```r
# the plot() function has a type argument, by default, it i a scatter plot, if you set it to"l" it is a line chart

# Line chart, defaukt set to l
plot(education.high$high, type="l")

# default scatter chart
plot(education.high$high)
```

```r
# Set it to “h” and you get a high density chart
# High-density
plot(education.high$high, type="h")
```

```r
# Set it to “s” and you get a step chart
# Step
plot(education.high$high, type="s")
```

```r
#?plot       to get more docs
```

```r
# Working with more arguments
# Set las to 1 to change the label positions so that they are horizontal.
plot(education.high$high, las=1)
```

```r
# Labelling Charts
plot(education.high$high, las=1, xlab="States", ylab="Percent", main="At least high school degree or equivalent by state")
```

```r
# You can also set the size of shapes (cex), labels (cex.axis), remove border (bty), or change the symbols used (pch).

plot(education.high$high, las=1, xlab="States", ylab="Percent", main="At
least high school degree or equivalent", bty="n", cex=0.5, cex.axis=0.6,pch=19)
```

## Additional charts

```r
# Bar plot
barplot(education$high)
```

```r
# Bar plot with changed parameters
barplot(education$high, names.arg=education$state, horiz=TRUE, las=1,
cex.names=0.5, border=NA)
```

```r
# Single box plot
boxplot(education$high)
```

```r
# Multiple box plots for comparison
boxplot(education[,2:4])
```

```r
# ####################################################################
# Multiple charts in one window
#    Finally, you can set some universal parameters so that you don’t have to specify them in every function.
#    For example, mfrow sets the number of rows and columns if you want to show multiple charts in the
#    same window, and mar sets the margins in between charts. Then like in previous examples, you’ve seen
#    las and bty, but now you don’t have to plug it in everywhere.
# ####################################################################
```

```r
# Multiple charts in one window
par(mfrow=c(3,3), mar=c(2,5,2,1), las=1, bty="n")
plot(education.high$high)
plot(education$high, education$bs)
plot(education.high$high, type="l") # Line
plot(education.high$high, type="h") # High-density
plot(education.high$high, type="s") # Step
barplot(education$high)
barplot(education$high, names.arg=education$state, horiz=TRUE, las=1, cex.names=0.5, border=NA)
boxplot(education$high)
boxplot(education[,2:4])
```
